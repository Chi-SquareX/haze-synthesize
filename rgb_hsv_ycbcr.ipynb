```python
%pylab inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
```

```python
img_source = "./imgs/holien.jpg"
img = mpimg.imread(img_source)
imgplot = plt.imshow(img)
plt.show()
```

```python
image = cv2.imread(img_source)
```

## RGB to YCbCr

```python
imgYCC = cv2.cvtColor(image, cv2.COLOR_BGR2YCR_CB)
plt.imshow(imgYCC)
plt.show()
```

```python
y = imgYCC[:,:,0]
cb = imgYCC[:,:,1]
cr = imgYCC[:,:,2] 
```

```python
plt.imshow(y)
plt.show()
```

```python
plt.imshow(cb)
plt.show()
```

```python
plt.imshow(cr)
plt.show()
```

```python
cv2.imwrite("y.jpg", y)
cv2.imwrite("cb.jpg", cb)
cv2.imwrite("cr.jpg", cr)
```

## RGB to HSV

```python
imgHSV = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
plt.imshow(imgHSV)
plt.show()
```

```python
h = imgHSV[:,:,0]
s = imgHSV[:,:,1]
v = imgHSV[:,:,2] 
```

```python
plt.imshow(h)
plt.show()
```

```python
plt.imshow(s)
plt.show()
```

```python
plt.imshow(v)
plt.show()
```

```python
imgGRAY = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(imgGRAY)
plt.show()
```

```python
cv2.imwrite("a.jpg", imgGRAY)
```
